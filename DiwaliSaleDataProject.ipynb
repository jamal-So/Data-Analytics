# Explotatory Data Analysis
### Gender

```python
ax = sns.countplot(x='Gender', data=df, palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python

# Group by 'Gender' and sum 'Amount'
sale_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set the style and color palette
sns.set_style('whitegrid')
palette = sns.color_palette('Set2')

# Create the bar plot
ax = sns.barplot(x='Gender', y='Amount', data=sale_gen, palette=palette)

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Add title and labels
ax.set_title('Total Amount by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Total Amount')

# Show plot
plt.show()

```

#### From the above graphs most of the buyers are female are purchasing power of the female is greater than the man]

# Age

```python
ax = sns.countplot(data = df,x ="Age Group", hue = "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
Sale_Age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = Sale_Age)
plt.xlabel= "Age Group"
plt.ylabel= "Amount"
```

##### From Above Graph we can see that most of the buyers are of age group between 26-35 yrs Female


# State

```python
# Total Number of Sale from 10 State
sale_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending =False).head(10)
sns.set(rc ={'figure.figsize':(16,6)})
sns.barplot(data=sale_state,x='State',y='Orders')
plt.xlabel='State'
plt.ylabel='Orders'
```

```python
# Total Number of Sale from 10 State
sale_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False).head(10)
sns.set(rc ={'figure.figsize':(16,6)})
sns.barplot(data=sale_state,x='State',y='Amount')
plt.xlabel='State'
plt.ylabel='Amount'
```

##### from the Above graph we can see that the most of the orders are from uttar pradesh, Maharasthra, & Karnataka Respectivaly
